```
def fill(col):
    c = col[0]
    if c==0 or c==6:
        return 5
    else:
        return c
    
df['EDUCATION'] = df[['EDUCATION']].apply(fill, axis=1)
```

```
def temp(col):
  c = col[0]
  if c==0:
    return 1;
  return c

df['MARRIAGE'] = df[['MARRIAGE']].apply(temp, axis=1)
```

```
df.drop(['ID'], axis=1, inplace=True)
```

```
df.head()
```

```
dataset = df
dataset = pd.get_dummies(dataset, columns = ['SEX', 'EDUCATION', 'MARRIAGE'])
dataset.head()
```

```
from sklearn.preprocessing import StandardScaler
stdSclr = StandardScaler()
col2scale = ['LIMIT_BAL',	'AGE',	'PAY_0',	'PAY_2',	'PAY_3',	'PAY_4',	'PAY_5',	'PAY_6', 	'BILL_AMT1',	'BILL_AMT2',	'BILL_AMT3',	'BILL_AMT4',	'BILL_AMT5',	'BILL_AMT6',	'PAY_AMT1',	'PAY_AMT2',	'PAY_AMT3',	'PAY_AMT4',	'PAY_AMT5',	'PAY_AMT6']
dataset[col2scale] = stdSclr.fit_transform(dataset[col2scale])
dataset.head()
```

```
from sklearn.model_selection import train_test_split
y = dataset['default_payment_next_month']
X = dataset.drop(['default_payment_next_month'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=0)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr = LogisticRegression(max_iter=2000)
lr.fit(X_train, y_train)

pred = lr.predict(X_test)
print('Logistic Regression accuracy score: ', accuracy_score(y_test, pred)*100)
```
